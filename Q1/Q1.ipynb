```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from math import pow
```

```python
def npv(d,m,g,r):
    if g > r*100:
        print("r must be larger than g", end='\n')
        return np.nan
    '''
    n0_5 = -5 * (1-pow((1-d/100)/(1+r),6)) / (1-((1-d/100)/(1+r)))
    n6_m = 0.5 / pow(1+r,6) * (1 - 1 / pow(1+r, m-5)) / (1 - 1 / (1+r))
    nm1 = 0.5 * (1 + (g/100)) / pow(1+r,m+1)
    n_inf = nm1 / (1 - (1+g/100)/(1+r))
    print(f'{n0_5} {n6_m} {n_inf}')
    '''
    n0_5 = -5 * (1 + r) / (r + d/100) * (1 - pow((1-d/100)/(1+r),6))
    n6_m = (0.5/pow(1+r,5)/r)*(1-pow(1/(1+r),m-5))
    n_inf = 0.5/(r-g/100)*(1+g/100)/pow(1+r, m)
    #print(f'{n0_5} {n6_m} {n_inf}')
    
    ans = n0_5 + n6_m + n_inf
    return ans
```

```python
print(npv(10,8,5,0.1))
```

```python
d = np.arange(0,50,5)
m = np.arange(0,50,1)
g = 5 #5$
r = 0.1


plt.title("change m(0 to 50)")
plt.xlabel("m")
plt.ylabel("NP (M)")
for td in d:
    pv = []
    for tm in m:
        pv = np.append(pv, npv(td,tm,g,r))
    plt.plot(m, pv, label = f'd = {td}')
plt.legend()
plt.show()
```
